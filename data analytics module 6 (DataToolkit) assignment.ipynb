```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

np.random.seed(30)

data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = go.Figure(data=[go.Scatter3d(
    x=df['X'],
    y=df['Y'],
    z=df['Z'],
    mode='markers',
    marker=dict(
        color=df['X'] + df['Y'] + df['Z'],
        colorscale='Viridis',
        opacity=0.8,
        size=5
    )
)])

fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        xaxis=dict(nticks=5),
        yaxis=dict(nticks=5),
        zaxis=dict(nticks=5)
    )
)

fig.show()
```

        np.random.seed(20)
        data={
        'Month': np.random.choice(['Jan','Feb','Mar', 'Apr','May',100),
        'Day': np.random.choice (range (1, 31), 100),
        'Sales': np.random.randint(1000 ,5000, 100)
        } 
        df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table to represent sales across months and days
pivot_df = df.pivot_table(values='Sales', index='Month', columns='Day', aggfunc='sum')

# Create a heatmap using Plotly
fig = px.imshow(pivot_df,
                labels=dict(x="Day", y="Month", color="Sales"),
                x=pivot_df.columns,
                y=pivot_df.index,
                color_continuous_scale='Viridis')

fig.update_layout(title='Sales Heatmap (Month vs. Day)')
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

y = np.linspace(-5, 5, 100)
x = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot of z=sin√(x^2+y^2)', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

# **BOKEH**

Ques 1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x, y, legend_label="Sine Wave", line_width=2)
show(p)
```

Ques 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.random.randint(5, 20, 100)
colors = np.random.choice(['red', 'green', 'blue'], 100)

data = dict(x=x, y=y, size=sizes, color=colors)
source = ColumnDataSource(data=data)

p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.scatter(x='x', y='y', size='size', fill_color='color', line_color='white', source=source)

show(p)
```

Ques 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
    fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
    counts = [20, 25, 30, 35]

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", 
           toolbar_location=None, tools="", 
           x_axis_label='Fruit', y_axis_label='Count')

p.vbar(x='fruits', top='counts', width=0.9, source=source, 
       legend_field="fruits", line_color='white', 
       fill_color="red")

show(p)
```

Ques 4. Create a Bokeh histogram to visualize the distribution of the given data.
    data_hist=np.random.randn(1000)
    hist, edges= np.histogram(data_hist,bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import  Range1d

np.random.seed(0)
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

midpoints = (edges[1:] + edges[:-1]) / 2

p = figure(title="Histogram", 
           tools="pan,wheel_zoom,reset", 
           x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="green", line_color="white")

p.x_range = Range1d(-4, 4)
show(p)
```
