Ans 3) Accessing items in list:
            samplelist = ["white", "red", "blue","pink"]
            print(samplelist[1])
            #Output: white
       Modifying items in list:
            samplelist = ["white", "red", "blue","pink"]
            samplelist[1] = "black"
            print(samplelist)
            #Output: ['white', 'black', 'blue', 'pink']
       Deleting items in list:
            samplelist = ["white", "red", "blue","pink"]
            samplelist.remove("red")
            print(samplelist)
            #Output: ['white','blue','pink]
