```python
# Ans 4)
def is_prime(n):
           if n <= 1:
               return False
           for i in range(2, int(n ** 0.5) + 1):
               if n % i == 0:
                   return False
           return True

num = int(input("Enter a number between 1 and 200: "))
       
if num < 1 or num > 200:
           print("Number is out of range. Please enter a number between 1 and 200.")
elif is_prime(num):
           print(f"{num} is a prime number")
else:
           print(f"{num} is not a prime number")
```

```python
# Ans 5)
       
class FibonacciIterator:

              def __init__(self, n):
                 self.n = n
                 self.a, self.b = 0, 1
                 self.count = 0
       
              def __iter__(self):
                 return self
       
              def __next__(self):
                  if self.count < self.n:
                     result = self.a
                     self.a, self.b = self.b, self.a + self.b
                     self.count += 1
                     return result
                  else:
                      raise StopIteration
            
num_digits = int(input("Enter the number of digits in Fibonacci series: "))
       
fib_iter = FibonacciIterator(num_digits)
       
print("Fibonacci series with", num_digits, "digits:")
for num in fib_iter:
       print(num)
```

```python
# Ans 6)
def powers_of_two(exponent):
         for i in range(exponent + 1):
              yield 2 ** i

exponent = int(input("Enter the exponent value: "))

print("Powers of 2 up to exponent", exponent, ":")
for power in powers_of_two(exponent):
        print(power)
```

```python
# Ans 7)
def read_file_line_by_line(file_path):
         with open(file_path, 'r') as file:
            for line in file:
               yield line.strip()

file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
           print(line)
```

```python
# Ans 9)
def celsius_to_fahrenheit(celsius):
           return (celsius * 9/5) + 32
celsius_temps = [15,44,50,12,88]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))
print(fahrenheit_temps)
```

```python
# Ans 10)
def is_vowel(char):
           return char.lower() in 'aeiou'
input_string = input("Enter a string: ")
no_vowels = ''.join(filter(lambda char: not is_vowel(char), input_string))
print("String without vowels:", no_vowels)
```
