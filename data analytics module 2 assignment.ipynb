		5//2 #output=2

		2<5 #output=True

		x = 50
 		y = 98
		print(x > 8 or y < 20)  # Output: True

		a = 5  # 0b0101

Ans7) There are 2 types of loops in python:
	a)for loop-The for loop in Python is used to iterate over a sequence (example as a list, tuple, dictionary,set,or string) or other iterable objects. It is useful when you want to execute a block of code a specific number of times or iterate over items in a collection.
	Example-
		for i in range(0,10)
		    print ("*",end="")
		#output=**********
	
	b)while loop-The while loop in Python repeatedly executes a block of code as long as a condition is True. It is useful when the number of iterations is not known beforehand and depends on some condition.
	Example-
		count = 0
                while count < 10:
    		   print("*",end="")
    		count += 1
		output:**********
		
	(loops can be used within loops that condition is called nested loop)
